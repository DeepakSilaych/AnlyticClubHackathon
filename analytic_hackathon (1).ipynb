```python
kproto = KPrototypes(n_clusters=2, verbose=2, max_iter=20)
cluster = kproto.fit_predict(newarray, categorical=[1,2,3,4,6,7])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
```

```python
numerical_features = ['age', 'stud_h']
categorical_features = ['year', 'sex', 'glang', 'part', 'job', 'health', 'psyt']
```

```python
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(drop='first', sparse=False)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('kmeans', KMeans(n_clusters=2, random_state=42))
])
```

```python
pipeline.fit(df)
```

```python
df['cluster'] = pipeline.named_steps['kmeans'].labels_
```

```python
plt.scatter(df['job'], df['stud_h'], c=df['cluster'], cmap='viridis')
plt.title('Data points and their clusters')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import seaborn as sns
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(30, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
false_df = df[df['cluster']==0]
true_df = df[df['cluster']==1]
```

```python
correlation_matrix = true_df.corr()
plt.figure(figsize=(30, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation_matrix = false_df.corr()
plt.figure(figsize=(30, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
numerical_summary = pd.DataFrame({
    'Cluster 0 Mean Age': [false_df['age'].mean()],
    'Cluster 1 Mean Age': [true_df['age'].mean()],
    'Cluster 0 Mean Stud_h': [false_df['stud_h'].mean()],
    'Cluster 1 Mean Stud_h': [true_df['stud_h'].mean()],
})
```

```python
categorical_summary = pd.DataFrame()
for col in categorical_features:
    false_df_mode = false_df[col].mode().values[0]
    true_df_mode = true_df[col].mode().values[0]
    categorical_summary[f'Cluster 0 Mode {col}'] = [false_df_mode]
    categorical_summary[f'Cluster 1 Mode {col}'] = [true_df_mode]
```

```python
cluster_summary = pd.concat([numerical_summary, categorical_summary], axis=1)
```
